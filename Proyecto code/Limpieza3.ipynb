```python

def separarendoscolumnas(df):
    df['Fecha'] = df['Fecha y hora'].str.split(' ', expand=True)[0]
    df['Hora'] = df['Fecha y hora'].str.split(' ', expand=True)[0]       
    df.drop(['Fecha y hora'], axis=1, inplace=True)
    df.drop(['Unnamed: 0'], axis=1, inplace=True)

    return df
```
